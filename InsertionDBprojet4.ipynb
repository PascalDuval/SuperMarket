```python
# %% Imports
import os
import sqlite3
from datetime import datetime, timedelta
import pandas as pd
from pathlib import Path

# %% Chemins et connexions
DB_PATH = Path("database/db/SuperMarket1.db")
XLSX_PATH = Path("PourInsertionBD.xlsx")

print(f"📂 CWD       : {Path.cwd().resolve()}")
print(f"📄 DB        : {DB_PATH.resolve()}")
print(f"📊 Excel OK? : {XLSX_PATH.exists()}")

conn = sqlite3.connect(DB_PATH)
conn.execute("PRAGMA foreign_keys = ON;")
print("✅ Connexion SQLite ouverte")

# %% Création des tables
with conn:
    conn.executescript("""
    PRAGMA foreign_keys = ON;

    CREATE TABLE IF NOT EXISTS Produits (
        EAN TEXT PRIMARY KEY,
        categorie TEXT NOT NULL,
        Rayon TEXT NOT NULL,
        Libelle_produit TEXT NOT NULL,
        prix REAL NOT NULL CHECK (prix >= 0)
    );
    CREATE INDEX IF NOT EXISTS idx_prod_rayon     ON Produits(Rayon);
    CREATE INDEX IF NOT EXISTS idx_prod_categorie ON Produits(categorie);

    CREATE TABLE IF NOT EXISTS Clients (
        CUSTOMER_ID TEXT PRIMARY KEY,
        date_inscription TEXT NOT NULL
            CHECK (length(date_inscription) = 10 AND date_inscription GLOB '____-__-__')
    );
    CREATE INDEX IF NOT EXISTS idx_clients_date ON Clients(date_inscription);

    CREATE TABLE IF NOT EXISTS Employé (
        id_employe TEXT PRIMARY KEY
            CHECK (length(id_employe) = 32 AND id_employe GLOB '[0-9a-f]*'),
        employe TEXT UNIQUE NOT NULL,
        prenom TEXT NOT NULL,
        nom TEXT NOT NULL,
        date_debut TEXT NOT NULL
            CHECK (length(date_debut) = 10 AND date_debut GLOB '____-__-__'),
        hash_mdp TEXT,
        mail TEXT UNIQUE NOT NULL
    );

    CREATE TABLE IF NOT EXISTS "Vente Détail" (
        ID_BDD TEXT PRIMARY KEY,
        CUSTOMER_ID TEXT NOT NULL,
        id_employe TEXT NOT NULL,
        EAN TEXT NOT NULL,
        Date_achat TEXT NOT NULL
            CHECK (length(Date_achat) = 10 AND Date_achat GLOB '____-__-__'),
        ID_ticket TEXT NOT NULL,
        FOREIGN KEY (CUSTOMER_ID) REFERENCES Clients(CUSTOMER_ID)
            ON UPDATE CASCADE ON DELETE RESTRICT,
        FOREIGN KEY (id_employe) REFERENCES Employé(id_employe)
            ON UPDATE CASCADE ON DELETE RESTRICT,
        FOREIGN KEY (EAN) REFERENCES Produits(EAN)
            ON UPDATE CASCADE ON DELETE RESTRICT
    );

    CREATE INDEX IF NOT EXISTS idx_vd_date    ON "Vente Détail"(Date_achat);
    CREATE INDEX IF NOT EXISTS idx_vd_client  ON "Vente Détail"(CUSTOMER_ID);
    CREATE INDEX IF NOT EXISTS idx_vd_ean     ON "Vente Détail"(EAN);
    CREATE INDEX IF NOT EXISTS idx_vd_employe ON "Vente Détail"(id_employe);
    CREATE INDEX IF NOT EXISTS idx_vd_ticket  ON "Vente Détail"(ID_ticket);
    """)
print("✅ Tables créées ou vérifiées")

# %% Fonctions utilitaires
EXCEL_BASE = datetime(1899, 12, 30)

def excel_to_iso(val):
    if pd.isna(val): return None
    if isinstance(val, (datetime, pd.Timestamp)):
        return val.strftime("%Y-%m-%d")
    try:
        return (EXCEL_BASE + timedelta(days=int(float(val)))).strftime("%Y-%m-%d")
    except:
        for fmt in ("%Y-%m-%d", "%d/%m/%Y", "%d/%m/%y"):
            try:
                return datetime.strptime(str(val).strip(), fmt).strftime("%Y-%m-%d")
            except:
                continue
        return str(val).strip()

def clean_text(val):
    return str(val).strip() if not pd.isna(val) else None

# %% Chargement Excel
assert XLSX_PATH.exists(), f"Fichier Excel manquant : {XLSX_PATH}"
xls = pd.ExcelFile(XLSX_PATH)
print(f"✅ Excel chargé : {XLSX_PATH}")
print("Feuilles disponibles :", xls.sheet_names)

# %% Insertion Produits
df = pd.read_excel(xls, "Produits")
df = df.assign(
    EAN=df["EAN"].apply(clean_text),
    categorie=df["categorie"].astype(str).str.strip(),
    Rayon=df["Rayon"].astype(str).str.strip(),
    Libelle_produit=df["Libelle_produit"].astype(str).str.strip(),
    prix=pd.to_numeric(df["prix"], errors="coerce")
)
rows = df.to_sql("Produits", conn, if_exists="append", index=False)
print(f"✅ Produits insérés : {rows}")

# %% Insertion Clients
df = pd.read_excel(xls, "Clients")
col_id = "CUSTOMER_ID" if "CUSTOMER_ID" in df.columns else "CUSTUMER_ID"
df = pd.DataFrame({
    "CUSTOMER_ID": df[col_id].apply(clean_text),
    "date_inscription": df["date_inscription"].apply(excel_to_iso)
})
rows = df.to_sql("Clients", conn, if_exists="append", index=False)
print(f"✅ Clients insérés : {rows}")

# %% Insertion Employé
df = pd.read_excel(xls, "Employé")
df = df.assign(
    id_employe=df["id_employe"].astype(str).str.strip().str.lower(),
    employe=df["employe"].astype(str).str.strip(),
    prenom=df["prenom"].astype(str).str.strip(),
    nom=df["nom"].astype(str).str.strip(),
    date_debut=df["date_debut"].apply(excel_to_iso),
    hash_mdp=df["hash_mdp"].astype(str).str.strip(),
    mail=df["mail"].astype(str).str.strip()
)
rows = df.to_sql("Employé", conn, if_exists="append", index=False)
print(f"✅ Employés insérés : {rows}")

# %% Insertion Vente Détail
df = pd.read_excel(xls, "Vente Détail")
df = df.assign(
    ID_BDD=df["ID_BDD"].apply(clean_text),
    CUSTOMER_ID=df["CUSTOMER_ID"].apply(clean_text),
    id_employe=df["id_employe"].astype(str).str.strip().str.lower(),
    EAN=df["EAN"].apply(clean_text),
    Date_achat=df["Date achat"].apply(excel_to_iso),
    ID_ticket=df["ID ticket"].apply(clean_text)
)
rows = df.to_sql("Vente Détail", conn, if_exists="append", index=False)
print(f"✅ Vente Détail insérées : {rows}")

# %% Vérifications
def count_rows(table):
    return conn.execute(f'SELECT COUNT(*) FROM "{table}"').fetchone()[0]

print("📊 Volumes :")
for table in ["Produits", "Clients", "Employé", "Vente Détail"]:
    print(f" - {table:13}: {count_rows(table)} lignes")

# CA du 2024-08-14
sql = """
SELECT v.Date_achat, ROUND(SUM(p.prix), 2) AS CA
FROM "Vente Détail" v
JOIN Produits p ON p.EAN = v.EAN
WHERE v.Date_achat = ?
GROUP BY v.Date_achat;
"""
date_test = "2024-08-14"
result = conn.execute(sql, (date_test,)).fetchone()
print(f"💶 CA du {date_test} :", result[1] if result else "0.00")

# Contrôle FK
checks = {
    "Ventes sans client":  'SELECT COUNT(*) FROM "Vente Détail" v LEFT JOIN Clients  c ON c.CUSTOMER_ID = v.CUSTOMER_ID WHERE c.CUSTOMER_ID IS NULL;',
    "Ventes sans employé": 'SELECT COUNT(*) FROM "Vente Détail" v LEFT JOIN Employé  e ON e.id_employe   = v.id_employe   WHERE e.id_employe IS NULL;',
    "Ventes sans produit": 'SELECT COUNT(*) FROM "Vente Détail" v LEFT JOIN Produits p ON p.EAN          = v.EAN          WHERE p.EAN IS NULL;'
}
for label, sql in checks.items():
    n = conn.execute(sql).fetchone()[0]
    print(f"🔍 {label:<25}: {n}")
```

```python
# %% [markdown]
# # Insertion dans la base existante : Clients, Employé, Vente Détail

# %% Imports
import sqlite3
import pandas as pd
from datetime import datetime, timedelta
from pathlib import Path

# %% Connexion à la base existante
DB_PATH = Path("database/db/SuperMarket1.db")
XLSX_PATH = Path("PourInsertionBD.xlsx")

conn = sqlite3.connect(DB_PATH)
conn.execute("PRAGMA foreign_keys = ON;")
print("✅ Connexion à la base existante réussie")

# %% Fonctions utilitaires

EXCEL_BASE = datetime(1899, 12, 30)

def excel_to_iso(val):
    if pd.isna(val): return None
    if isinstance(val, (datetime, pd.Timestamp)):
        return val.strftime("%Y-%m-%d")
    try:
        return (EXCEL_BASE + timedelta(days=int(float(val)))).strftime("%Y-%m-%d")
    except:
        for fmt in ("%Y-%m-%d", "%d/%m/%Y", "%d/%m/%y"):
            try:
                return datetime.strptime(str(val).strip(), fmt).strftime("%Y-%m-%d")
            except:
                continue
        return None

def clean_text(val):
    return str(val).strip() if not pd.isna(val) else None

# %% Chargement Excel
xls = pd.ExcelFile(XLSX_PATH)
print("📊 Feuilles Excel :", xls.sheet_names)

# %% Insertion Clients (avec nettoyage et filtrage des dates invalides)

df = pd.read_excel(xls, "Clients")
col_id = "CUSTOMER_ID" if "CUSTOMER_ID" in df.columns else "CUSTUMER_ID"

df_clean = pd.DataFrame({
    "CUSTOMER_ID": df[col_id].apply(clean_text),
    "date_inscription": df["date_inscription"].apply(excel_to_iso)
})

# Filtre les dates valides
mask_valid = df_clean["date_inscription"].str.match(r"^\d{4}-\d{2}-\d{2}$")
df_valid = df_clean[mask_valid & df_clean["CUSTOMER_ID"].notna()]

# Log des lignes rejetées
rejected = df_clean[~mask_valid | df_clean["CUSTOMER_ID"].isna()]
if not rejected.empty:
    print("⚠️ Lignes Clients rejetées :")
    display(rejected)

# Insertion
rows = df_valid.to_sql("Clients", conn, if_exists="append", index=False)
print(f"✅ Clients insérés : {rows}")

# %% Insertion Employé

df = pd.read_excel(xls, "Employé")
df = df.assign(
    id_employe=df["id_employe"].astype(str).str.strip().str.lower(),
    employe=df["employe"].astype(str).str.strip(),
    prenom=df["prenom"].astype(str).str.strip(),
    nom=df["nom"].astype(str).str.strip(),
    date_debut=df["date_debut"].apply(excel_to_iso),
    hash_mdp=df["hash_mdp"].astype(str).str.strip(),
    mail=df["mail"].astype(str).str.strip()
)

# Filtre dates valides
df = df[df["date_debut"].str.match(r"^\d{4}-\d{2}-\d{2}$")]
rows = df.to_sql("Employé", conn, if_exists="append", index=False)
print(f"✅ Employés insérés : {rows}")

# %% Insertion Vente Détail

df = pd.read_excel(xls, "Vente Détail")
df = df.assign(
    ID_BDD=df["ID_BDD"].apply(clean_text),
    CUSTOMER_ID=df["CUSTOMER_ID"].apply(clean_text),
    id_employe=df["id_employe"].astype(str).str.strip().str.lower(),
    EAN=df["EAN"].apply(clean_text),
    Date_achat=df["Date achat"].apply(excel_to_iso),
    ID_ticket=df["ID ticket"].apply(clean_text)
)

# Filtrage
df = df[df["Date_achat"].str.match(r"^\d{4}-\d{2}-\d{2}$")]
rows = df.to_sql("Vente Détail", conn, if_exists="append", index=False)
print(f"✅ Ventes insérées : {rows}")

# %% Vérification des volumes
def count_rows(tbl):
    return conn.execute(f'SELECT COUNT(*) FROM "{tbl}"').fetchone()[0]

for table in ["Clients", "Employé", "Vente Détail"]:
    print(f"{table:<15}: {count_rows(table)} lignes")

# %% Contrôles FK
checks = {
    "Ventes sans client":  'SELECT COUNT(*) FROM "Vente Détail" v LEFT JOIN Clients  c ON c.CUSTOMER_ID = v.CUSTOMER_ID WHERE c.CUSTOMER_ID IS NULL;',
    "Ventes sans employé": 'SELECT COUNT(*) FROM "Vente Détail" v LEFT JOIN Employé  e ON e.id_employe   = v.id_employe   WHERE e.id_employe IS NULL;',
    "Ventes sans produit": 'SELECT COUNT(*) FROM "Vente Détail" v LEFT JOIN Produits p ON p.EAN          = v.EAN          WHERE p.EAN IS NULL;'
}
for label, sql in checks.items():
    n = conn.execute(sql).fetchone()[0]
    print(f"🔍 {label:<25}: {n}")

# %% Fin
conn.close()
print("🔒 Connexion fermée")
```

```python
# %% Insertion Clients (version robuste)

import re

# Lecture
df = pd.read_excel(xls, "Clients")
col_id = "CUSTOMER_ID" if "CUSTOMER_ID" in df.columns else "CUSTUMER_ID"

# Nettoyage
df_clean = pd.DataFrame({
    "CUSTOMER_ID": df[col_id].apply(clean_text),
    "date_inscription": df["date_inscription"].apply(excel_to_iso)
})

# Validation stricte : format ISO AAAA-MM-JJ
def is_valid_iso_date(date_str):
    return bool(re.fullmatch(r"\d{4}-\d{2}-\d{2}", str(date_str)))

mask_valid = df_clean["date_inscription"].apply(is_valid_iso_date)
mask_notna = df_clean["CUSTOMER_ID"].notna()
df_valid = df_clean[mask_valid & mask_notna]

# Lignes rejetées (affichées pour inspection)
rejected = df_clean[~(mask_valid & mask_notna)]
if not rejected.empty:
    print("⚠️ Lignes invalides rejetées (Clients) :")
    display(rejected)

# Insertion
rows = df_valid.to_sql("Clients", conn, if_exists="append", index=False)
print(f"✅ Clients insérés : {rows}")
```
