```python
import pandas as pd
import sqlite3

# Connexion à la base SQLite
db_path = "database/db/SuperMarketOlap.db"
conn = sqlite3.connect(db_path)

# Charger le CSV des faits
df_faits = pd.read_csv("data/csv/Faits_Ventes.csv")

# Charger les dimensions existantes
dim_clients   = pd.read_sql("SELECT CUSTOMER_ID FROM Dim_Clients", conn)
dim_produits  = pd.read_sql("SELECT EAN FROM Dim_Produits", conn)
dim_employe   = pd.read_sql("SELECT id_employe FROM Dim_Employe", conn)
dim_dates     = pd.read_sql("SELECT date_excell FROM Dim_Calendrier", conn)

# Vérifier les correspondances
valid_faits = df_faits[
    df_faits["CUSTOMER_ID"].isin(dim_clients["CUSTOMER_ID"]) &
    df_faits["EAN"].isin(dim_produits["EAN"]) &
    df_faits["id_employe"].isin(dim_employe["id_employe"]) &
    df_faits["Date_Achat"].isin(dim_dates["date_excell"])
]

# Lignes invalides
invalid_faits = df_faits.loc[~df_faits.index.isin(valid_faits.index)]

print(f"✅ Faits valides : {len(valid_faits)} lignes")
print(f"⚠️ Faits invalides (non insérés) : {len(invalid_faits)} lignes")

if not invalid_faits.empty:
    print("\nExemple de faits invalides :")
    print(invalid_faits.head())

# Importer seulement les lignes valides
valid_faits.to_sql("Faits_Ventes", conn, if_exists="append", index=False)

conn.close()
print("🎉 Import terminé")
```

```python
import pandas as pd
import sqlite3

# Connexion SQLite
conn = sqlite3.connect("database/db/SuperMarketOlap.db")

# Charger le CSV complet
df_faits = pd.read_csv("data/csv/Faits_Ventes.csv")

# Charger les dimensions
dim_clients   = pd.read_sql("SELECT CUSTOMER_ID FROM Dim_Clients", conn)["CUSTOMER_ID"].astype(str).str.strip().str.upper()
dim_produits  = pd.read_sql("SELECT EAN FROM Dim_Produits", conn)["EAN"].astype(str).str.strip()
dim_employe   = pd.read_sql("SELECT id_employe FROM Dim_Employe", conn)["id_employe"].astype(str).str.strip()
dim_dates     = pd.read_sql("SELECT date_excell FROM Dim_Calendrier", conn)["date_excell"].astype(str).str.strip()

# Normaliser les faits
df_faits["CUSTOMER_ID"] = df_faits["CUSTOMER_ID"].astype(str).str.strip().str.upper()
df_faits["EAN"] = df_faits["EAN"].astype(str).str.strip()
df_faits["id_employe"] = df_faits["id_employe"].astype(str).str.strip()
df_faits["Date_Achat"] = df_faits["Date_Achat"].astype(str).str.strip()

# Marquer les incohérences
df_faits["valid"] = (
    df_faits["CUSTOMER_ID"].isin(dim_clients) &
    df_faits["EAN"].isin(dim_produits) &
    df_faits["id_employe"].isin(dim_employe) &
    df_faits["Date_Achat"].isin(dim_dates)
)

# Séparer
valid_faits = df_faits[df_faits["valid"]].drop(columns=["valid"])
invalid_faits = df_faits[~df_faits["valid"]].drop(columns=["valid"])

print(f"✅ Faits valides : {len(valid_faits)}")
print(f"❌ Faits invalides : {len(invalid_faits)}")

# Sauvegarder pour inspection
valid_faits.to_csv("data/csv/Faits_Ventes_valid.csv", index=False)
invalid_faits.to_csv("data/csv/Faits_Ventes_invalid.csv", index=False)

conn.close()
```

```python
import pandas as pd
import sqlite3

# Connexion SQLite
conn = sqlite3.connect("database/db/SuperMarketOlap.db")

# Charger les logs
df_logs = pd.read_csv("data/csv/Logs.csv")

def ca_depuis_logs(date_excel):
    """Retourne le CA calculé à partir des logs pour une date donnée"""
    df_sel = df_logs[
        (df_logs["date"] == date_excel) &
        (df_logs["action"].str.upper() == "INSERT") &
        (df_logs["table_insert"].str.lower().str.contains("ventes"))
    ]
    
    ids = df_sel["id_ligne"].astype(str).unique()
    if len(ids) == 0:
        return 0.0
    
    query = f"""
    SELECT SUM(p.prix) AS ca
    FROM Faits_Ventes f
    JOIN Dim_Produits p ON f.EAN = p.EAN
    WHERE f.ID_BDD IN ({','.join(['?']*len(ids))});
    """
    df_ca = pd.read_sql(query, conn, params=list(ids))
    return float(df_ca.iloc[0,0] or 0.0)

# Calcul du CA du 14 et du 15 août
ca_14 = ca_depuis_logs(45518)
ca_15 = ca_depuis_logs(45519)
ca_total = ca_14 + ca_15

print(f"💰 CA logué le 14/08/2024 : {ca_14:,.2f} €".replace(",", " ").replace(".", ","))
print(f"💰 CA logué le 15/08/2024 : {ca_15:,.2f} €".replace(",", " ").replace(".", ","))
print(f"📊 CA cumulé (14 + 15 août) : {ca_total:,.2f} €".replace(",", " ").replace(".", ","))

conn.close()
```

```python


# Charger les logs
df_logs = pd.read_csv("data/csv/Logs.csv")

# Convertir colonne date en numérique
df_logs["date"] = pd.to_numeric(df_logs["date"], errors="coerce")

# Comptage du nombre total d'opérations par date
ops_par_date = df_logs.groupby("date").size().reset_index(name="nb_operations")

# Extraire les opérations du 14 (45518) et du 15 (45519)
ops_14 = int(ops_par_date.loc[ops_par_date["date"] == 45518, "nb_operations"].sum())
ops_15 = int(ops_par_date.loc[ops_par_date["date"] == 45519, "nb_operations"].sum())
total_ops = ops_14 + ops_15

# Calcul des pourcentages
pct_14 = (ops_14 / total_ops) * 100 if total_ops > 0 else 0
pct_15 = (ops_15 / total_ops) * 100 if total_ops > 0 else 0

print("=== Récapitulatif des opérations ===")
print(f"📅 14 août (45518) : {ops_14:,} opérations ({pct_14:.2f}%)".replace(",", " "))
print(f"📅 15 août (45519) : {ops_15:,} opérations ({pct_15:.2f}%)".replace(",", " "))
print(f"🔢 Total analysé    : {total_ops:,} opérations".replace(",", " "))
```
